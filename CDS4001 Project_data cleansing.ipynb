```python
import pandas as pd
```

```python
yelp_ds = pd.read_csv("Yelp dataset.csv", encoding='unicode_escape')
```

```python
yelp_ds
```

```python
yelp_metads = pd.read_csv("Yelp Metadata.csv")
```

```python
yelp_metads
```

```python
yelp_merged = yelp_metads.merge(yelp_ds)
yelp_merged
```

```python
missing_values = yelp_merged.isnull()
missing_values
#No missing values
```

```python
#size of df
yelp_merged.shape
```

```python
#data cleansing
#remove null data and duplicated data
yelp_merged = yelp_merged.dropna().drop_duplicates()

#remove the useless column: Review_id, Product_id, Review_Date since it is descriptive data when the data is taken only
columns_to_remove = ['Review_id', 'Product_id', 'Review_Date']
yelp_merged = yelp_merged.drop(columns=columns_to_remove)
yelp_merged
```

```python
type(yelp_merged.columns[1])
```

```python
correlation = yelp_merged['Label'].corr(yelp_merged['Rating'])
correlation
```

```python
yelp_merged_types = yelp_merged.dtypes
yelp_merged_types
```

```python
yelp_merged['Label'].value_counts()

#it implies the dataset is imbalance between two results
```

```python
yelp_merged_positive = yelp_merged[yelp_merged['Label'] == 1]
yelp_merged_positive
```

```python
yelp_merged_negative = yelp_merged[yelp_merged['Label'] == -1]
yelp_merged_negative
```

```python
#227981 records in a dataset seems too large.
#It is because it is a bit time comsuming to analyse and my computer may not be affordable to run due to its RAM.
#Thus it would be better to randomly select 5000 records to a new dataset.



#In order to be a balnaced new dataframe
#Take 2500 records from positive and negative result dataframe respectively
new_yelp_positive = yelp_merged_positive.sample(n=1000, random_state=42)
new_yelp_negative = yelp_merged_negative.sample(n=1000, random_state=42)

#Then combine them together
new_yelp_df = pd.concat([new_yelp_positive, new_yelp_negative], axis=0)

#Randomly shuffle the records within the DataFrame
new_yelp_df = new_yelp_df.sample(frac=1, random_state=42)

# Reset the index of the shuffled DataFrame
new_yelp_df = new_yelp_df.reset_index(drop=True)
new_yelp_df
```

```python
new_yelp_df['Label'].value_counts()
```
